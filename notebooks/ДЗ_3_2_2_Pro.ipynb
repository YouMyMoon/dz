## Задание 1

**Условие:**
Создайте случайный целочисленный массив ```ndarray``` из ```1000``` элементов, принадлежащих интервалу ```[2,10]```. С помощью булевой индексации найдите все простые элементы в массиве. Найдите сумму полученных элементов. Используйте ```np.random.seed(1000)``` перед генерацией массива для воспроизводимости результатов.

**Формат ответа:**
Число (например, ```94```)



**Решение:**

````python
import numpy as np

# Установка seed для воспроизводимости результатов
np.random.seed(1000)

# Создание случайного целочисленного массива ndarray из 1000 элементов на интервале [2,10]
array = np.random.randint(2, 11, size=1000)

# Вывод созданного массива
print("Массив:")
print(array)

# Функция для определения простого числа
def is_prime(n):
    if n <= 1:
        return False
    elif n <= 3:
        return True
    elif n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# Булевая индексация для поиска всех простых элементов
prime_elements = array[np.array([is_prime(x) for x in array])]

# Нахождение суммы простых элементов
sum_prime_elements = np.sum(prime_elements)

# Вывод простых элементов и их суммы
print("\nПростые элементы:")
print(prime_elements)
print("\nСумма простых элементов:", sum_prime_elements)
````

## Задание 2

**Условие:**
Создайте случайный целочисленный массив ```ndarray``` размера ```(10,10)``` элементов, принадлежащих отрезку ```[0,10]```. Найдите сумму квадратов элементов главной диагонали, не используя циклы. Используйте ```np.random.seed(1000)``` перед генерацией массива для воспроизводимости результатов.

**Формат ответа:**
Число (например, ```150```)



**Решение:**

````python
import numpy as np

# Установка seed для воспроизводимости результатов
np.random.seed(1000)

# Создание случайного целочисленного массива ndarray размера (10,10) элементов на отрезке [0,10]
array = np.random.randint(0, 11, size=(10, 10))

# Вывод созданного массива
print("Массив:")
print(array)

# Нахождение главной диагонали с помощью np.diag
main_diagonal = np.diag(array)

# Нахождение суммы квадратов элементов главной диагонали с помощью np.sum и np.square
sum_of_squares = np.sum(np.square(main_diagonal))

# Вывод суммы квадратов элементов главной диагонали
print("\nСумма квадратов элементов главной диагонали:", sum_of_squares)
````

## Задание 3

**Условие:**
Создайте случайный трехмерный целочисленный массив ```ndarray``` размера ```(10,10,10)``` элементов, принадлежащих отрезку ```[0,10]```. Напомним, что подвыборка по первой оси вида ```[0,:,:]``` является матрицей размера ```(10,10)```. Найти поэлементную сумму всех матриц, полученных описанной выше операцией среза при прохождении вдоль всей первой оси (```[0,:,:]```, ```[1,:,:]``` и т.д.). В ответе укажите максимальный элемент в получившейся матрице. Используйте ```np.random.seed(1000)``` перед генерацией массива для воспроизводимости результатов.

**Формат ответа:**
Число (например, ```50```)



**Решение:**

````python
import numpy as np

# Установка seed для воспроизводимости результатов
np.random.seed(1000)

# Создание случайного трехмерного целочисленного массива ndarray размера (10,10,10) элементов на отрезке [0,10]
array = np.random.randint(0, 11, size=(10, 10, 10))

# Вывод созданного массива
print("Массив:")
print(array)

# Выполнение суммирования по первой оси с помощью np.sum
sum_along_first_axis = np.sum(array, axis=0)

# Вывод результата суммирования
print("\nСумма всех матриц по первой оси:")
print(sum_along_first_axis)

# Нахождение максимального элемента в получившейся матрице
max_element = np.max(sum_along_first_axis)

# Вывод максимального элемента
print("\nМаксимальный элемент в получившейся матрице:", max_element)
````

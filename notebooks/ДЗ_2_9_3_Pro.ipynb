## Задание 1

**Условие:**
Перевести текст ```_Text_``` в нижний регистр(```str.lower```), подсчитать статистику встречаемости символов(модуль ```collections```), получить лист всех слов(модуль ```re```).



**Решение:**

````python
# Не изменять!
text = r'Представим, что какая-нибудь высокоразвитая цивилизация нашла быстровращающуюся чёрную дыру и построила вокруг неё сферическую оболочку из зеркал, обращённых внутрь. Оболочку сплошную, как каноническая сфера Дайсона. Благо, чёрные дыры намного меньше звёзд, так что и оболочку построить несравненно проще. Теперь открываем в ней отверстие и запускаем внутрь пучок электромагнитных волн. Эти волны получают с помощью эргосферы ускорение и вылетают наружу, отражаются от зеркала, возвращаются в эргосферу, ещё больше ускоряются, опять вылетают, отражаются, возвращаются, ускоряются, и т. д. (какая-то часть волн будет потеряна из-за падения на горизонт событий). Каждое попадание излучения в эргосферу приводит его экспоненциальному усилению. Это так называемый эффект рассеивания с помощью сверхизлучения, впервые предсказанный советским физиком Яковом Зельдовичем.'
````

````python
import collections
import re



# Переводим текст в нижний регистр
lower_text = text.lower()

# Подсчитываем статистику встречаемости символов
character_counts = collections.Counter(lower_text)

# Выводим статистику встречаемости символов
print("Статистика встречаемости символов:")
print(character_counts)

# Получаем список всех слов с помощью регулярного выражения
words_list = re.findall(r'\b\w+\b', lower_text)

# Выводим список слов
print("\nСписок всех слов:")
print(words_list)
````

## Задание 2

**Условие:**
Напишите функцию, которая просит пользователя ввести название параграфа и страницу, после чего разбирает данную строку и делает красивый форматированный вывод в виде: ```Название параграфа---------12```. Количество знаков ```-``` зависит от ширины поля, заданного в параметрах функции.  



**Решение:**

````python
def format_paragraph(paragraph_name, page_number, field_width=30):
    # Форматируем строку с названием параграфа и номером страницы
    formatted_string = f"{paragraph_name.ljust(field_width, '-')} {str(page_number).rjust(2)}"
    # Выводим отформатированную строку
    print(formatted_string)

# Запрашиваем у пользователя название параграфа и номер страницы
paragraph = input("Введите название параграфа: ")
page = int(input("Введите номер страницы: "))

# Вызываем функцию для форматирования и вывода информации
format_paragraph(paragraph, page)
````
